```python
import pandas as pd
import numpy as np
import os
```

```python
res_dir = '/scratch/zz737/fiar/tournaments/results/tournament_1_extension_1/'
```

```python
def combine_one_dir(res_dir):

    fn_base = lambda x:f'round_robin_{x}.pkl'
    results_all = []
    max_i = len(os.listdir(res_dir))

    for i in range(max_i):
        fn = fn_base(i)
        try:
            result_one = pd.read_pickle(os.path.join(res_dir, fn))
            results_all.append(result_one)
        except:
            pass

    results_all = pd.concat(results_all,axis=0)

#     results_all = results_all.set_index('Unnamed: 0')

    combined_fn = os.path.join(res_dir, fn_base('combined'))
    results_all.to_pickle(combined_fn)
    print('Combined at: '+combined_fn)
```

```python
combine_one_dir(res_dir)
```
